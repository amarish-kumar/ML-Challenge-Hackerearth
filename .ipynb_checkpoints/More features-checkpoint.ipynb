```python
def LabelEncoding(dataset, column_index):
    l = pp.LabelEncoder()
    l.fit(dataset[:, column_index])
    var = l.transform(dataset[:, column_index])
    return var 
```

```python
#4th column, term
term = LabelEncoding(ds, 4)
ts_term = LabelEncoding(ts, 4)

# print np.unique(term, return_counts=True)

#7th column, grade
grade = LabelEncoding(ds, 7)
ts_grade = LabelEncoding(ts, 7)

#8th column, sub_grade
sub_grade = LabelEncoding(ds, 8)
ts_sub_grade = LabelEncoding(ts, 8)

#11th column, home_ownership
home_ownership = LabelEncoding(ds, 11)
ts_home_ownership = LabelEncoding(ts, 11)

#13th column, verification_status
verification_status = LabelEncoding(ds, 13)
ts_verification_status = LabelEncoding(ts, 13)

#14th column, pymnt_plan
pymnt_plan = LabelEncoding(ds, 14)
ts_pymnt_plan = LabelEncoding(ts, 14)

#16th column, purpose
purpose = LabelEncoding(ds, 16)
ts_purpose = LabelEncoding(ts, 16)

#18th column, zip_code
zip_code = LabelEncoding(ds, 18)
ts_zip_code = LabelEncoding(ts, 18)

#19th column, addr_state
addr_state = LabelEncoding(ds, 19)
ts_addr_state = LabelEncoding(ts, 19)

#30th column, initial_list_status
initial_list_status = LabelEncoding(ds, 30)
ts_initial_list_status = LabelEncoding(ts, 30)

#37th column, application_type
application_type = LabelEncoding(ds, 37)
ts_application_type = LabelEncoding(ts, 37)

# print np.unique(term, return_counts=True)
```

```python
def Impute(dataset, column_index):
    da = dataset[:, column_index].reshape(1, -1)
    l = pp.Imputer(strategy='mean', axis=1)
    var = l.fit_transform(da)
    var = var[0]
    return var 
```

```python
annual_inc = Impute(ds, 12)
delinq_2yrs = Impute(ds, 21)
inq_last_6mths = Impute(ds, 22)
mths_since_last_delinq = Impute(ds, 23)
mths_since_last_record = Impute(ds, 24)
open_acc = Impute(ds, 25)
pub_rec = Impute(ds, 26)
revol_util = Impute(ds, 28)
total_acc = Impute(ds, 29)
collections_12_mths_ex_med = Impute(ds, 35)
mths_since_last_major_derog = Impute(ds, 36)
acc_now_delinq = Impute(ds, 40)
tot_coll_amt = Impute(ds, 41)
tot_cur_bal = Impute(ds, 42)
total_rev_hi_lim = Impute(ds, 43)

ts_annual_inc = Impute(ts, 12)
ts_delinq_2yrs = Impute(ts, 21)
ts_inq_last_6mths = Impute(ts, 22)
ts_mths_since_last_delinq = Impute(ts, 23)
ts_mths_since_last_record = Impute(ts, 24)
ts_open_acc = Impute(ts, 25)
ts_pub_rec = Impute(ts, 26)
ts_revol_util = Impute(ts, 28)
ts_total_acc = Impute(ts, 29)
ts_collections_12_mths_ex_med = Impute(ts, 35)
ts_mths_since_last_major_derog = Impute(ts, 36)
ts_acc_now_delinq = Impute(ts, 40)
ts_tot_coll_amt = Impute(ts, 41)
ts_tot_cur_bal = Impute(ts, 42)
ts_total_rev_hi_lim = Impute(ts, 43)
```

```python
import keras
from keras.layers import Activation, Dense
from keras.models import Sequential
from keras.utils.np_utils import to_categorical
```

```python
nn_label = to_categorical(label)
nn_y_train = nn_label[:split]
nn_y_test = nn_label[split:]
```

```python
model = Sequential()
model.add(Dense(256, input_shape=(22,), activation='tanh'))
model.add(Dense(128, activation='tanh'))
model.add(Dense(64, activation='tanh'))
model.add(Dense(2))
model.add(Activation('softmax'))
model.summary()
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
hist = model.fit(x_train, nn_y_train, shuffle=True, batch_size=256, nb_epoch=10, validation_data=(x_test, nn_y_test))
```
